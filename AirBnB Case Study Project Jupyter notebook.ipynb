```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../AirBnB/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

neighborhood = air_bnb.groupby('neighbourhood_group').count().sort_values('id', ascending =False)
print("5 groups and Manhattan appears the most")
neighborhood
```

```python
# Are private rooms the most popular in manhattan?

most_popular = air_bnb[air_bnb["neighbourhood_group"]=='Manhattan'].groupby('room_type').count().sort_values("id", ascending = False)

print("No, entire home/apt is")
most_popular
```

```python
# Which hosts are the busiest and based on their reviews?
busiest_host = air_bnb.groupby(["host_name"]).sum('number_of_Reviews').sort_values('number_of_reviews', ascending=False)
print("Michael is the busiest based on number of reviews")
busiest_host
```

```python
#Which neighorhood group has the highest average price?
high_avg = air_bnb.groupby(["neighbourhood_group"]).agg({"price": [np.size, np.mean]})
print("manhattan has the highest price")
high_avg
```

```python
# Which neighbor hood group has the highest total price?
high_total_price = air_bnb.groupby(["neighbourhood_group"]).sum('price').sort_values('price', ascending = False)
print("Manhattan has the highest total price")
high_total_price
```

```python
#Which top 5 hosts have the highest total price?
high_host_price = air_bnb.groupby(["host_name"]).sum('price').sort_values('price', ascending = False).head(5)
print("Sonder, Blueground, Micheal, David, Alex")
high_host_price
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
no_availability = air_bnb[(air_bnb.availability_365==0)& (air_bnb.number_of_reviews> 100)].groupby('host_name').sum().sort_values('number_of_reviews', ascending = False)
no_availability
```

```python
# What host has the highest total of prices and where are they located?

high_prices = air_bnb.groupby(['host_name','neighbourhood_group']).sum("price").sort_values("price", ascending= False)
print("Sonder in Manhattan")
high_prices
```

```python
# When did Danielle from Queens last receive a review?

danielle_queens = air_bnb[(air_bnb.host_name =="Danielle")& (air_bnb.neighbourhood_group =="Queens")].sort_values("last_review", ascending = False)[["last_review"]]
print("2019-07-08")
danielle_queens
```
